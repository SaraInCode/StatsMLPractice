```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
label=pd.DataFrame(list(map(lambda x: iris.target_names[x],iris.target)),columns=['Species'])
iris=pd.concat([data,label],axis=1)
print(iris.head())
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1)
sns.distplot(iris['sepal length (cm)'],color='r')
plt.subplot(2,2,2)
sns.distplot(iris['sepal width (cm)'],color='g')
plt.subplot(2,2,3)
sns.distplot(iris['petal length (cm)'])
plt.subplot(2,2,4)
sns.distplot(iris['petal width (cm)'])
```

```python
sns.boxplot(data=iris, palette="Set2")
```

```python
sns.countplot(x='Species',data=iris)
```

```python
sns.pairplot(iris,hue_order=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)','Species'],palette='coolwarm')
```

```python
sns.lmplot('sepal length (cm)','petal length (cm)',data=iris,hue='Species')
```

```python
sns.barplot(x='Species',y='sepal length (cm)',data=iris)
```

```python
sns.heatmap(iris.corr(),cmap='coolwarm')
plt.title('iris.corr()')
```
