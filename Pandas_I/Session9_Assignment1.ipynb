```python
import pandas as pd
baby_names=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
```

```python
baby_names.info()
```

## 1. Delete unnamed columns

```python
del baby_names['Unnamed: 0']
```

```python
baby_names.info()
```

## 2. Distribution of male and female

```python
baby_names['Gender'].value_counts()
```

## 3. Top 5 most preferred names

```python
del baby_names["Id"]
del baby_names["Year"]
sort1 = baby_names.groupby("Name").sum()
sort1.head()
print(sort1.shape)
sort1.sort_values("Count", ascending = 0).head()
```

## 4. Median name occurence

```python
sort1[sort1.Count == sort1.Count.median()]
```

## 5. Male and female born count by states

```python
sort2 = baby_names.groupby(["State","Gender"]).sum()
sort2.head()
print(sort2.shape)
sort2.sort_values("Count", ascending = 0).head()
```
